```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# Data
data = [
    ['thermax', 'थरमैक्स', 'थर्मैक्स'],
    ['sikhaaega', 'सिखाएगा', 'सिखागे'],
    ['learn', 'लर्न', 'ल्यर्न'],
    ['twitters', 'ट्विटर्स', 'ट्विटर्स'],
    ['tirunelveli', 'तिरुनेलवेली', 'तिरुनेवलेली'],
    ['independence', 'इंडिपेंडेंस', 'इंडेडेंसंस'],
    ['speshiyon', 'स्पेशियों', 'स्पेशियों'],
    ['shurooh', 'शुरूः', 'शुरूह'],
    ['kolhapur', 'कोल्हापुर', 'कोल्हापुर'],
    ['ajhar', 'अजहर', 'अझर'],
    ['karaar', 'क़रार', 'करार'],
    ['udhed', 'उधेड़', 'उधेड']
]

# Prepare the plot
fig, ax = plt.subplots()

# Create a colorful grid
colors = ['red', 'green', 'blue']
for i, (input_val, output_val, predicted_val) in enumerate(data):
    rect = plt.Rectangle((i % 3, i // 3), 1, 1, facecolor=colors[i % 3], alpha=0.7)
    ax.add_patch(rect)
    
    font_prop = FontProperties(fname='/content/NotoSansDevanagari-VariableFont_wdth,wght.ttf')  # Replace with the path to your Hindi font file
    plt.text(i % 3 + 0.5, i // 3 + 0.5, f'{input_val}\n{output_val}\n{predicted_val}',
             ha='center', va='center', color='white', fontsize=10, fontproperties=font_prop)

# Set the axis limits and labels
ax.set_xlim(0, 3)
ax.set_ylim(0, 4)
ax.set_xticks([])
ax.set_yticks([])

# Set the plot title
plt.title('Input / Output / Predicted Output Grid', fontsize=12)

# Display the plot
plt.show()
```
